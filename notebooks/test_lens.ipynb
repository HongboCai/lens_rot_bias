```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import healpy as hp
from pixell import curvedsky
import matplotlib as mpl
mpl.rcParams['font.size']=13
```

```python
ls, clpp = np.loadtxt('../inputPs/cosmo2017_10K_acc3_lenspotentialCls.dat', usecols=(0,5), unpack=True)
ls = np.concatenate(([0,1],ls))
clpp = np.concatenate(([0,0],clpp))
```

```python
factor = 0.5*ls**2
```

```python
phi_alm = curvedsky.rand_alm(clpp, lmax=6000, seed=(0,0,2,0))
kappa_alm = hp.almxfl(phi_alm, factor)
```

```python
clpp_sim = hp.alm2cl(phi_alm)
clkk_sim = hp.alm2cl(kappa_alm)
```

```python
plt.plot(ls[:3000], clpp_sim[:3000])
plt.xlabel('L')
plt.ylabel('$C_L^{\phi\phi}$')
plt.legend()
plt.xscale('log')
plt.yscale('log')
```

```python
plt.plot(ls[:3000], clkk_sim[:3000])
plt.xlabel('L')
plt.ylabel('$C_L^{\kappa\kappa}$')
plt.legend()
plt.xscale('log')
plt.yscale('log')
```
